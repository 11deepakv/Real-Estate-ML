Using SimpleImputer

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='median')
X = imputer.fit_transform(data)
```

```python
data = pd.DataFrame(X, columns=data.columns)
data.describe()
```

# Checking Correlations

```python
corr = data.corr()
corr['MEDV'].sort_values(ascending=False)
```

```python
from pandas.plotting import scatter_matrix
attributes = ['MEDV', 'RM', 'TAX', 'LSTAT']
scatter_matrix(data[attributes], figsize=(12, 9))
```

# Adding Combinations

```python
data['TAXPRM'] = data['TAX']/data['RM']
data.head(5)
```

```python
corr = data.corr()
corr['MEDV'].sort_values(ascending = False)
```

```python
data = data.drop('TAX', axis=1)
```

# Train  Test splitting

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(data, data['CHAS']):
    strat_train_set = data.iloc[train_index]
    strat_test_set = data.iloc[test_index]
```

```python
strat_train_set.describe()
```

```python
strat_test_set.describe()
```

```python
strat_train_set['CHAS'].value_counts()
```

```python
strat_test_set['CHAS'].value_counts()
```

```python
data = strat_train_set.copy()
```

```python
data_feature = data.drop(['MEDV'], axis=1)
```

```python
data_feature
```

```python
data_labels = data['MEDV'].copy()
```

```python
data_labels
```

```python
corr = data.corr()
corr['MEDV'].sort_values(ascending=False)
```

# Creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('standard_scalr', StandardScaler())
])
```

```python
data_prep = my_pipeline.fit_transform(data_feature)
data_prep.shape
```

```python
data_prep
```

# Selecting a Desired Model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
# model = LinearRegression()
# model = DecisionTreeRegressor()
model = RandomForestRegressor()
model.fit(data_prep, data_labels)
```

```python
rand_feature = data_feature.iloc[:10]
rand_labels = data_labels.iloc[:10]
prep_data = my_pipeline.transform(rand_feature)
predications = model.predict(prep_data)
predications
```

```python
list(rand_labels)
```

# Evaluating the model

```python
from sklearn.metrics import mean_squared_error
import numpy as np
predication = model.predict(data_prep)
mse = mean_squared_error(data_labels, predication)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```

# Using Better Evaluation Technique - Cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, data_prep, data_labels, scoring='neg_mean_squared_error', cv = 10)
-scores
```

```python
rmse_scores = np.sqrt(-scores)
rmse_scores
```

```python
def print_scores(scores):
    print('Scores: ', scores)
    print('Mean: ', scores.mean())
    print('STD: ', scores.std())
```

```python
print_scores(rmse_scores)
```

# Launching Model

```python
from joblib import dump, load
dump(model, 'Reat_Estate')
```

# Testing model on test Data

```python
x_test = strat_test_set.drop('MEDV', axis=1)
y_test = strat_test_set['MEDV'].copy()
prep_x_test = my_pipeline.transform(x_test)
final_pred = model.predict(prep_x_test)
mse = mean_squared_error(y_test, final_pred)
mse
```

```python
rmse = np.sqrt(mse)
rmse
```
